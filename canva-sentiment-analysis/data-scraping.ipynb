Dengan mengakses data review aplikasi ini dari **Google Play Store**,  dapat diketahui beragam pendapat pengguna yang bisa digunakan untuk analisis sentimen.

```python
import pandas as pd
from google_play_scraper import app, reviews, Sort, reviews_all
import csv
```

## **Scraping Data**

```python
reviewcanva = reviews_all(
    'com.canva.editor',  
    lang='id',        
    country='id',          
    sort=Sort.MOST_RELEVANT,
    count=15000      
)
```

```python
with open('review_canva.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Review'])  # Menulis header kolom
    for review in reviewcanva:
        writer.writerow([review['content']])  # Menulis konten ulasan ke dalam file CSV
```

```python
df = pd.DataFrame(reviewcanva)
```
